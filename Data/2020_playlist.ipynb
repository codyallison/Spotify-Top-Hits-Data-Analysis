```python
import requests
from config import client_id
from config import client_secret
from formulas import get_playlist_songs,get_track_features, get_spotify_token, get_album_data
import pandas as pd
```

```python
playlist = get_playlist_songs("37i9dQZF1DX7Jl5KP2eZaS")
playlist
```

```python
song_list = []
for x in range(0,50):
    song_list.append(playlist['tracks']['items'][x]['track']['name'])
song_list
```

```python
song_id_list = []
for x in range(0,50):
    song_id_list.append(playlist['tracks']['items'][x]['track']['id'])
song_id_list
```

```python
song_popularity =[]
for x in range(0,50):
    song_popularity.append(playlist['tracks']['items'][x]['track']['popularity'])
song_popularity
```

```python
song_artist =[]
for x in range(0,50):
    song_artist.append(playlist['tracks']['items'][x]['track']['artists'][0]['name'])
song_artist
```

```python
tracks_df_20 = pd.DataFrame({"Song Title":song_list,
                            "Artist": song_artist,
                          "Song ID":song_id_list,
                           "popularity": song_popularity})
tracks_df_20["danceability"] = ""
tracks_df_20["energy"] = ""
tracks_df_20["key"] = ""
tracks_df_20["loudness"] = ""
tracks_df_20["mode"] = ""
tracks_df_20["speechiness"] = ""
tracks_df_20["acousticness"] = ""
tracks_df_20["instrumentalness"] = ""
tracks_df_20["liveness"] = ""
tracks_df_20["valence"] = ""
tracks_df_20["tempo"] = ""
tracks_df_20["duration_ms"] = ""
tracks_df_20
```

```python
for index, row in tracks_df_20.iterrows():
    
    song_id = tracks_df_20.loc[index, "Song ID"]
    song_features = get_track_features(song_id)
    
    tracks_df_20.loc[index, 'danceability'] = song_features['danceability']
    tracks_df_20.loc[index, 'energy'] = song_features['energy']
    tracks_df_20.loc[index, 'key'] = song_features['key']
    tracks_df_20.loc[index, 'loudness'] = song_features['loudness']
    tracks_df_20.loc[index, 'mode'] = song_features['mode']
    tracks_df_20.loc[index, 'speechiness'] = song_features['speechiness']
    tracks_df_20.loc[index, 'acousticness'] = song_features['acousticness']
    tracks_df_20.loc[index, 'instrumentalness'] = song_features['instrumentalness']
    tracks_df_20.loc[index, 'liveness'] = song_features['liveness']
    tracks_df_20.loc[index, 'valence'] = song_features['valence']
    tracks_df_20.loc[index, 'tempo'] = song_features['tempo']
    tracks_df_20.loc[index, 'duration_ms'] = song_features['duration_ms']
```

```python
tracks_df_20
```

```python
tracks_df_20 = tracks_df_20.dropna(how='any')
```

```python
tracks_df_20 =tracks_df_20.rename(columns = {'duration_ms': 'duration (ms)', 'tempo':'bpm'})
                              
tracks_df_20['duration (secs)'] = tracks_df_20['duration (ms)']/1000
tracks_df_20
```

```python
tracks_df_20['top year'] = '2020'
tracks_df_20
```

```python
tracks_df_20.to_csv("2020_2022_outputs/2020_output.csv")
```
