```python
import requests
from config import client_id
from config import client_secret
from formulas import get_playlist_songs,get_track_features, get_spotify_token
import pandas as pd
```

```python
#playlist for 2022 top hits https://open.spotify.com/playlist/37i9dQZF1DX18jTM2l2fJY?si=edfbbcae36c1476f
playlist = get_playlist_songs("37i9dQZF1DX18jTM2l2fJY")
playlist
```

```python
song_list = []
for x in range(0,50):
    song_list.append(playlist['tracks']['items'][x]['track']['name'])
song_list
```

```python
song_id_list = []
for x in range(0,50):
    song_id_list.append(playlist['tracks']['items'][x]['track']['id'])
song_id_list
```

```python
song_popularity =[]
for x in range(0,50):
    song_popularity.append(playlist['tracks']['items'][x]['track']['popularity'])
song_popularity
```

```python
song_artist =[]
for x in range(0,50):
    song_artist.append(playlist['tracks']['items'][x]['track']['artists'][0]['name'])
song_artist
```

```python
tracks_df_22 = pd.DataFrame({"Song Title":song_list,
                            "Artist": song_artist,
                          "Song ID":song_id_list,
                           "popularity": song_popularity})
tracks_df_22["danceability"] = ""
tracks_df_22["energy"] = ""
tracks_df_22["key"] = ""
tracks_df_22["loudness"] = ""
tracks_df_22["mode"] = ""
tracks_df_22["speechiness"] = ""
tracks_df_22["acousticness"] = ""
tracks_df_22["instrumentalness"] = ""
tracks_df_22["liveness"] = ""
tracks_df_22["valence"] = ""
tracks_df_22["tempo"] = ""
tracks_df_22["duration_ms"] = ""
tracks_df_22
```

```python
for index, row in tracks_df_22.iterrows():
    
    song_id = tracks_df_22.loc[index, "Song ID"]
    song_features = get_track_features(song_id)
    
    tracks_df_22.loc[index, 'danceability'] = song_features['danceability']
    tracks_df_22.loc[index, 'energy'] = song_features['energy']
    tracks_df_22.loc[index, 'key'] = song_features['key']
    tracks_df_22.loc[index, 'loudness'] = song_features['loudness']
    tracks_df_22.loc[index, 'mode'] = song_features['mode']
    tracks_df_22.loc[index, 'speechiness'] = song_features['speechiness']
    tracks_df_22.loc[index, 'acousticness'] = song_features['acousticness']
    tracks_df_22.loc[index, 'instrumentalness'] = song_features['instrumentalness']
    tracks_df_22.loc[index, 'liveness'] = song_features['liveness']
    tracks_df_22.loc[index, 'valence'] = song_features['valence']
    tracks_df_22.loc[index, 'tempo'] = song_features['tempo']
    tracks_df_22.loc[index, 'duration_ms'] = song_features['duration_ms']
```

```python
tracks_df_22
```

```python
tracks_df_22 =tracks_df_22.rename(columns = {'duration_ms': 'duration (ms)','tempo':'bpm'})
                              
tracks_df_22['duration (secs)'] = tracks_df_22['duration (ms)']/1000
tracks_df_22
```

```python
tracks_df_22['top year'] = '2022'
tracks_df_22
```

```python
tracks_df_22.to_csv("2020_2022_outputs/2022_output.csv")
```
